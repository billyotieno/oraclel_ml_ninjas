```python
gas_dataset = pd.read_csv('customer_gas_survey.csv')
gas_dataset.head()
```

```python
gas_dataset.columns
```

```python
gas_dataset.drop(columns=['Row Id'], inplace=True)
```

```python
gas_dataset.head()
```

```python
new_row = gas_dataset.Location.str.split(" ", n = 4, expand=True)
new_row.head()
```

```python
gas_dataset["longitude"] = new_row[0]
gas_dataset["latitude"] = new_row[1]
gas_dataset["altitude"] = new_row[2]
gas_dataset["elevation"] = new_row[4]
gas_dataset
```

```python
gas_dataset.columns
```

```python
gas_dataset
```

```python
gas_dataset.drop(columns=['Completed', 'Completed By', 'Started', 'Received', 'Completed At',
       'TIMEANDDATE', 'Location'], inplace=True)
gas_dataset
```

```python
gas_dataset.isnull().sum()
```

```python
mapping_df = gas_dataset[["Address","Street","howoftendoyoubuygas","doyougetgasdeliveredorpickup","doesthebranofgasmatter","longitude","latitude"]]
```

```python
mapping_df.head()
```

```python
mapping_df.isna().sum()
```

```python
mapping_df.dropna(axis=0, inplace=True)
```

```python
mapping_df.head()
```

```python
mapping_df.dtypes
```

```python
mapping_df.longitude = mapping_df.longitude.astype("float64")
mapping_df.latitude = mapping_df.latitude.astype("float64")
```

```python
address = 'Kenya'

geolocator = Nominatim(user_agent="kenya_explore")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Kenya are {}, {}.'.format(latitude, longitude))
```

```python
map_kenya = folium.Map(location=[latitude, longitude], zoom_start=6)

# add markers to map
for lat, lng, label in zip(mapping_df['longitude'], mapping_df['latitude'], mapping_df['Address']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7).add_to(map_kenya)  
    
map_kenya
```
